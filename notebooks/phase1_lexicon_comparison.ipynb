# Phase 1: Lexicon Comparison

This notebook is an orchestration layer for the reusable Phase 1 pipeline in `src/lexicon_baselines.py` and the exploratory analysis helpers in `src/phase1_exploration.py`.

The exploration now shows both before and after duplicate removal so the experiment can assess whether duplicate reviews changed the rating and label distributions.

Label mapping used throughout the experiment:
- 1-2 stars -> Negative
- 3 stars -> Neutral
- 4-5 stars -> Positive

```python
from src.data_prep import load_amazon_reviews, prepare_and_save_dataset, prepare_dataset, resolve_raw_data_path
from src.lexicon_baselines import run_lexicon_models, save_phase1_outputs
from src.phase1_exploration import save_phase1_exploration_outputs
from src.utils import write_json
from src.config import METRICS_DIR
```

```python
source_path = resolve_raw_data_path()
raw_loaded_df = load_amazon_reviews(source_path)
before_duplicate_removal_df = prepare_dataset(raw_loaded_df, remove_exact_duplicates=False)
prepared_df, profile, paths = prepare_and_save_dataset(source_path)
before_duplicate_removal_df.shape, prepared_df.shape, paths
```

```python
save_phase1_exploration_outputs(before_duplicate_removal_df, prepared_df)
results_df, summary_df, metrics_by_model = run_lexicon_models(prepared_df)
save_phase1_outputs(results_df, summary_df, metrics_by_model)
write_json(profile, METRICS_DIR / 'dataset_profile.json')
summary_df
```
